#3-1 Money Change
In this problem, you will design and implement an elementary greedy algorithm
used by cashiers all over the world millions of times per day.

```python
%%writefile change.cpp

#include <iostream>

int get_change(int m)
{
    int a=0;
    int n=0;
    a=m/10;
    n=n+a;
    m=m-(a*10);

    a=m/5;
    n=n+a;
    m=m-(a*5);

    a=m/1;
    n=n+a;
    m=m-a;
    return n;
}

int main() {
  int m;
  std::cin >> m;
  std::cout << get_change(m) << '\n';
}
```

#3-3 Car Fueling
You are going to travel to another city that is located ð‘‘ miles away from your home city. Your car can travel
at most ð‘š miles on a full tank and you start with a full tank. Along your way, there are gas stations at
distances stop1, stop2, . . . , stopð‘› from your home city. What is the minimum number of refills needed?

```python
%%writefile car_fueling.cpp

#include <iostream>
#include <vector>

using std::cin;
using std::cout;
using std::vector;
using std::max;

int compute_min_refills(int dist, int tank, vector<int>stops, int n) {

    int para=0;
    int car=tank;
    if(dist<car)
    {
        return 0;
    }
    else
    {
        for(int i=0;i<n;i++)
        {
            if(stops[i]>car)
            {
                if(i!=0)
                {
                    car=stops[i-1]+tank;
                    para++;
                    if(car<stops[i])
                        return -1;
                }
                else{
                    return -1;
                }
            }
            if(i==n-1)
            {
                if(car<dist)
                {
                    car=stops[i]+tank;
                    para++;
                    if(car<dist)
                        return -1;
                }
            }
        }
    }
    return para;
}


int main() {
    int d = 0;
    cin >> d;
    int m = 0;
    cin >> m;
    int n = 0;
    cin >> n;

    vector<int> stops(n);
    for (size_t i = 0; i < n; ++i)
        cin >> stops.at(i);

    cout << compute_min_refills(d, m, stops, n) << '\n';

    return 0;
}
```

#3-6 Maximum Number of Prizes
You are organizing a funny competition for children. As a prize fund you have ð‘›
candies. You would like to use these candies for top ð‘˜ places in a competition
with a natural restriction that a higher place gets a larger number of candies.
To make as many children happy as possible, you are going to find the largest
value of ð‘˜ for which it is possible.

```python
%%writefile different_summands.cpp

#include <iostream>
#include <vector>

using std::vector;

vector<int> optimal_summands(int n) {
  vector<int> summands;
  int i=0;
  int a=1;
  int maxp=0;
  int dif=-1;

  while(i==0)
  {
      if((a*(a+1)/2)>=n)
      {
          maxp=a;
          i=1;
      }
      a++;
  }

  dif=(maxp*(maxp+1)/2)-n;

  if(dif==0)
  {
      summands.resize(maxp);
      for(int j=1;j<=maxp;j++)
      {
          summands[j-1]=j;
      }
  }
  else
  {
      summands.resize(maxp-1);
      for(int j=1;j<=maxp;j++)
      {
          if(j==maxp-1)
          {
              summands[j-1]=(2*j+1)-dif;
              j++;
          }
          else
          {
              summands[j-1]=j;
          }
      }
  }
  return summands;
}

int main() {
  int n;
  std::cin >> n;
  vector<int> summands = optimal_summands(n);
  std::cout << summands.size() << '\n';
  for (size_t i = 0; i < summands.size(); ++i) {
    std::cout << summands[i] << ' ';
  }
}
```
